```python
import pandas as pd
import numpy as np
```

```python
df_retrieval = pd.read_pickle('../data/processed/exper_mouse/mouse_retrieval_sw_StandardScaler.pck')
df_retrieval
```

```python
df_retrieval['cell_type_detail'] = df_retrieval['cell_type']
df_retrieval['cell_type'] = df_retrieval['cell_type'].str.split('_', expand=True)[0]

df_retrieval.loc[df_retrieval.index.isin(np.where(df_retrieval['cell_type'].isin(['E12', 'E14', 'adult']))[0]), 'cell_type'] = 'HSC'
df_retrieval.loc[df_retrieval.index.isin(np.where(df_retrieval['cell_type'].isin(['celebral-cortex', 'CNS', 'brain']))[0]), 'cell_type'] = 'neuron'

```

```python
df_retrieval = df_retrieval.loc[df_retrieval['cell_type'].isin(['HSC','4cell','ICM','spleen','8cell','neuron','zygote','2cell','ESC'])].reset_index(drop=True)
df_retrieval = df_retrieval.iloc[:, :-1]
# df_retrieval = df_retrieval.loc[df_retrieval['cell_type'].isin(['HSC','4cell','ICM'])].reset_index(drop=True)
print(df_retrieval.shape)
```

```python
df_retrieval.to_pickle('../data/processed/exper_mouse/mouse_retrieval_cv_sw_StandardScaler.pck')
```
